```python
def tran(path):
    with open(path,'r') as fp:
        sequences=[]
        result=''
        for f in fp:
            if f[0]=='>':
                sequences.append(result)
                result=''
            else:result+=f.strip()
    return count_transition_transversion(sequences[1],result)

def TransTransvRatio(s1,s2): 
    transCount=0
    transvCount=0
    mismatch_pairs=[ ]
    for i in range(len(s1)):
     pairs=0
    if s1[i] == s2[i]:
        pairs +=1
    if s1[i] != s2[i]:  
        pairs=s1[i]+s2[i]
        mismatch_pairs.append(pairs)
    for pair in mismatch_pairs:

        if pair =="AG" or pair=="GA" or pair=="CT" or pair=="TC":
         transCount+=1
        if pair == "AT" or pair == "TA" or pair == "CG" or pair == "GC":
         transvCount+=1
    return(transCount/transvCount)

def count_transition_transversion(s,t):
    transversion={'A':['T','C'],'G':['T','C'],'T':['A','G'],'C':['A','G']}
    transition={'A':'G','G':'A','T':'C','C':'T'}
    n=len(s)
    return sum([1 if transition[s[i]]==t[i] else 0 for i in range(n)])/sum([1 if t[i] in transversion[s[i]] else 0 for i in range(n)])
print(tran(r'rosalind_tran.txt'))


```

```python


from Bio import SeqIO

file =open('rosalind_sseq.txt', 'r')
records = SeqIO.parse(file, 'fasta')
list_seq = [record.seq for record in records]
S_string1 = list_seq[0]
T_string2 = list_seq[1]

lists =list()
list_a = list()
for i in range(len(T_string2)):
    positions = list()
    for j in range(len(S_string1)):
        if T_string2[i] != S_string1[j]:
            continue
        if T_string2[i] == S_string1[j]:
            positions.append(j+1)
    lists.append(positions)
    list_a += positions


sortedlist = sorted(set(list_a))
mylen=len(sortedlist)
prints = []
j = 0
for i in range(mylen):
        if j < len(lists) and sortedlist[i] in lists[j]:
            prints.append(sortedlist[i])
            j += 1
print(' '.join(str(i) for i in prints))
```
