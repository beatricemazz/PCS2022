- FIBD (Mortal Fibonacci Rabbits)

```python
month=89
lifespan=16
f=[1]*100
for i in range(2,month):
  f[i]=f[i-1]+f[i-2]
  if i>= lifespan:
    f[i]-=f[(i-lifespan)-1]  
print(f[i])
```


- REVP (locating restriction sites)

```python
from Bio import SeqIO                           

record = SeqIO.read('rosalind_revp.txt', 'fasta')
my_onward_seq = str(record.seq)                       
reverse = str(record.seq.complement())          

for i in range(len(my_onward_seq)-4+1):                   
  j=4         
  while j<=13:
    if  my_onward_seq[i:i+j] == (reverse[i:i+j]) [::-1]:    
        
        if i + j < len(my_onward_seq):
           print(i+1,j)                
        else:
            print(i+1, len(my_onward_seq)- i)
            i += 1
            break
    j+= 1
    
```

- lcsm Finding a Shared Motif

```python
from Bio import SeqIO                      #cp
sequences = []                             
handle = open('rosalind_lcsm.txt', 'r')
for record in SeqIO.parse(handle, 'fasta'):
    sequence = []                          
    seq = ''                               
    for nt in record.seq:                  
        seq += nt                          
    sequences.append(seq)                  
handle.close()

srt_seq = sorted(sequences, key=len)     
short_seq = srt_seq[0]                   
comp_seq = srt_seq[1:]                   
motif = ''                               
for i in range(len(short_seq)):          
    for j in range(i, len(short_seq)):   
        m = short_seq[i:j + 1]           
        found = False                   
        for sequ in comp_seq:            
            if m in sequ:                
                found = True            
            else:                        
                found = False           
                break                   
        if found and len(m) > len(motif):
            motif = m                    
print(motif)
```

```python
f= open("rosalind_lgis.txt","r")
f=f.readlines()
p=list(map(int,f[1].split()))
def lgis(f):  #cp
    incr = []
    decr = []
    for (i,x) in enumerate(f):
        incr.append(max([incr[j] for (j, y) in enumerate(f[:i]) if y < x] or [[]], key=len) + [x])
    for (i,x) in enumerate(f):
        decr.append(max([decr[j] for (j, y) in enumerate(f[:i]) if y > x] or [[]], key=len) + [x])
    print(*max(incr, key=len))
    print(*max(decr, key=len))
print(lgis(p))
```
